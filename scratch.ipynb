```python
df1 = pd.read_sql('''
SELECT *
FROM movie_basics;
''',conn)
df2 = pd.read_csv('zippedData/bom.movie_gross.csv.gz')

merged_df = df1.merge(df2, how='inner', left_on = 'original_title', right_on = 'title')
merged_df.head(5)
```

```python
merged_df = merged_df.drop_duplicates(['primary_title'])
merged_df.head(5)
```

```python
# Remove unwanted characters from foreign_gross column
merged_df['foreign_gross'] = merged_df['foreign_gross'].str.replace(',', '')

#
merged_df.plot.hist(bins='studio','domestic_gross')
```

```python
tn['worldwide_gross'] = tn['worldwide_gross'].str.replace('$', '')
tn['worldwide_gross'] = tn['worldwide_gross'].str.replace(',', '')
tn['worldwide_gross'] = tn['worldwide_gross'].astype(float)
tn['production_budget'] = tn['production_budget'].str.replace('$', '')
tn['production_budget'] = tn['production_budget'].str.replace(',', '')
tn['production_budget'] = tn['production_budget'].astype(float)
tn.head()
tn['margin'] = (tn['worldwide_gross']/tn['production_budget'])
```

```python
tn.head()
```

```python
tn.dtypes
```

```python
merged2 = merged_df.merge(tn, how='inner', left_on='title', right_on='movie')
```

```python
indie_df = tn[(tn['production_budget'] < 2000000)]
```

```python
merged2.plot.scatter('runtime_minutes', 'margin')
```

```python
# Drop an outlier
merged2 = merged2[merged2['margin'] < 100]
```

```python
merged2.plot.scatter('runtime_minutes', 'margin', alpha=.1)
;
```

```python
ind_run = indie_df.merge(df1, how='inner', left_on='movie', right_on='original_title')
```

```python
ind_run.boxplot('runtime_minutes','start_year')
;
```

```python
ind_rat = indie_df.merge(tmdb, how='inner', left_on='movie', right_on='original_title')
```
